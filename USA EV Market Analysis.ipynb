```python
import matplotlib.pyplot as plt

# Group by year and sum EV Registrations  How has EV adoption changed in the USA over tim
ev_yearly = ev_df.groupby("year")["EV Registrations"].sum().reset_index()
plt.figure(figsize=(10, 5))
plt.plot(ev_yearly["year"], ev_yearly["EV Registrations"], marker='o', color='blue')
for x, y in zip(ev_yearly["year"], ev_yearly["EV Registrations"]):
    label = f"{round(y / 1_000_000, 1)}M" 
    plt.text(x, y, label, fontsize=9, ha='center', va='bottom')
plt.title("Total EV Registrations in the USA (Year-wise)")
plt.xlabel("Year")
plt.ylabel("EV Registrations")
plt.tight_layout()
plt.show()

# from 2021, the EV market experienced a boom in registration, and the year  2023, the came with a drastic high registration than 2026 to 2020.   
```

```python
import seaborn as sns                                             
ev_income = ev_df[['Per_Cap_Income', 'EV Share (%)']].dropna()

plt.figure(figsize=(10,5))
sns.regplot(x='Per_Cap_Income',y='EV Share (%)',data=ev_income, scatter_kws={"alpha": 0.6})
plt.title("Relationship between income level and EV share(%)")
plt.xlabel("per capita income (USD)")
plt.ylabel("EV Share (%)")
plt.show()

correlation = ev_income["Per_Cap_Income"].corr(ev_income["EV Share (%)"])
print(f"Correlation between Income and EV Share: {correlation:.2f}")

# A range of income levels from 20000 to 40000 do not prefer the EV; rather, you can say they cannot afford an EV.
# but above the 50000 income level, people are buying the EV, and it is also affordable for that category of people.
```

<div style="background: linear-gradient(to right, blue, purple); padding: 10px; border-radius: 15px;">
  <p style="font-size:27px; font-family: Arial, sans-serif; color: #FFffff;">
    <strong>Which state had the highest YoY growth in EV registrations?</strong>
  </p>
</div>


```python
top_state = ev_df.groupby(['year','state'])['EV Registrations'].sum().reset_index()
yoy_state = top_state.sort_values(by=['state','year'])
yoy_state = yoy_state[yoy_state['EV Registrations'] != 0]
yoy_state['YOY growth (%)'] = yoy_state.groupby('state')['EV Registrations'].pct_change() * 100
state_1 = yoy_state.dropna(subset=['YOY growth (%)'])
yoy = state_1.loc[state_1['YOY growth (%)'].idxmax()]
print("State with Highest YoY Growth in EV Registrations:")
print(yoy)
# Oklahoma experienced the highest year-over-year growth in EV registrations, with an increase of 208.33% in 2018.
```
